## Memisahkan data training dan data testing

Memisahkan data training dan data testing dalam preprocessing adalah suatu langkah yang dilakukan untuk mempersiapkan data sebelum dilakukan proses pembelajaran atau training pada model machine learning. Tujuan dari memisahkan data ini adalah untuk menghindari overfitting pada model yang dibangun dan memastikan bahwa model yang dibangun dapat bekerja dengan baik pada data yang belum pernah dilihat sebelumnya.

Data training adalah data yang digunakan untuk melatih model machine learning dan untuk menyesuaikan parameter atau bobot dari model tersebut agar dapat menghasilkan prediksi yang akurat. Sedangkan data testing adalah data yang digunakan untuk menguji kinerja dari model yang telah dilatih pada data training. Data testing tidak digunakan untuk melatih model, namun digunakan untuk menguji keakuratan prediksi dari model.

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df_imputed, test_size = 0.2, random_state = 0)
```

```python
print(train.shape)
```

```python
print(test.shape)
```

```python
#Seperate the target and independent variable
train_X = train.drop(columns = ['Species'], axis = 1)
train_Y = train['Species']

test_X = test.drop(columns = ['Species'], axis = 1)
test_Y = test['Species']
```

```python
print(train_X.shape, train_Y.shape)
```

```python
print(test_X.shape, test_Y.shape)
```

## KNN

Algoritma K-Nearest Neighbors (K-NN) adalah salah satu algoritma yang digunakan dalam machine learning untuk melakukan klasifikasi dan regresi pada data. Secara sederhana, algoritma K-NN bekerja dengan mencari k data terdekat dari suatu data yang akan diprediksi, kemudian melakukan voting untuk menentukan kelas atau nilai prediksi dari data tersebut.

Dalam algoritma K-NN, k merupakan parameter yang menentukan jumlah data terdekat yang akan diambil untuk melakukan voting. Jumlah k yang optimal dapat ditentukan dengan mencoba-coba dan melakukan evaluasi kinerja model pada data testing.

Untuk melakukan prediksi dengan algoritma K-NN, langkah-langkah yang dilakukan adalah sebagai berikut:

Hitung jarak antara data yang akan diprediksi dengan seluruh data pada data training.
Ambil k data terdekat berdasarkan jarak yang dihitung pada langkah pertama.
Lakukan voting pada k data terdekat untuk menentukan kelas atau nilai prediksi dari data yang akan diprediksi.
Hasil prediksi dapat digunakan untuk melakukan klasifikasi atau regresi pada data.

```python
train_Y
```

```python
import sklearn.linear_model
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(train_X, train_Y)
```

```python
neigh.score(test_X, test_Y)
```

```python
pred = neigh.predict(test_X)
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_Y, pred))
```

## Naive Bayes

Algoritma Naive Bayes adalah salah satu algoritma pembelajaran mesin yang digunakan untuk melakukan klasifikasi pada dataset yang diberikan. Algoritma ini berdasarkan pada teori probabilitas dan asumsi bahwa semua fitur yang ada pada dataset tersebut saling independen satu sama lain.

Dalam melakukan klasifikasi, Naive Bayes memanfaatkan nilai probabilitas dari setiap fitur yang ada pada dataset untuk memprediksi kelas pada setiap data yang belum diketahui. Algoritma ini menghitung nilai probabilitas posterior dari kelas pada setiap data yang akan diklasifikasikan, berdasarkan nilai probabilitas prior dan likelihood yang diperoleh dari data training.

Dalam Naive Bayes, terdapat beberapa jenis algoritma seperti Gaussian Naive Bayes, Multinomial Naive Bayes, dan Bernoulli Naive Bayes. Pada Gaussian Naive Bayes, diasumsikan bahwa setiap fitur memiliki distribusi normal, sementara pada Multinomial Naive Bayes, diasumsikan bahwa setiap fitur terdistribusi multinomial. Sedangkan pada Bernoulli Naive Bayes, diasumsikan bahwa setiap fitur terdistribusi bernoulli.

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(train_X, train_Y)
```

```python
pred = gnb.predict(test_X)
print(pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(test_Y, pred)
```

```python
print(accuracy)
```

```python
neigh.score(test_X, test_Y)
```

```python
print(classification_report(test_Y, pred))
```
